```python
import pandas as pd
corpus = pd.read_csv("shuffled_text_master.csv")

import re
bills = corpus['big_text']
new_bills = []
for bill in bills:
    new_bills.append(re.sub("\s[0-9][0-9]*", "", bill))
```

```python
corpus['text'] = new_bills
corpus
```
